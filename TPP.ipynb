# sentence Tokenization

```python
import nltk
from nltk.tokenize import sent_tokenize
```

```python
sen=""" a grammatically complete group of words 
containing a subject!! & a predicate that expresses a statement."""
```

```python
t=sent_tokenize(sen)
print(t)
```

# stop words

```python
from nltk.corpus import stopwords 
nltk.download('stopwords')
stop_words=set(stopwords.words('english'))
```

```python
print(stop_words)
```

```python
tokens=word_tokenize(sen)
filtered=[word for word in tokens if word not in stop_words]
filtered
```

# stemming

```python
from nltk.stem import PorterStemmer
```

```python
stemmer=PorterStemmer()
stemmed_words=[stemmer.stem(word) for word in filtered]
print(stemmed_words)
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
```

```python
import nltk
nltk.download('wordnet')
```

```python
lemmatizer=WordNetLemmatizer()
lemmatizer_words=[lemmatizer.lemmatize(word) for word in filtered]
print(lemmatizer_words)
```

# Lemmatization with POS

```python
from nltk.corpus import wordnet

lemmatized_words=[lemmatizer.lemmatize(word,pos='v') for word in filtered]
print(lemmatized_words)
```
