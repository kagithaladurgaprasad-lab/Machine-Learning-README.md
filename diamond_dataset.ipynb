**carat**: The weight of the diamond, where 1 carat equals 200mg.
**cut**: The quality of the cut (Fair, Good, Very Good, Premium, Ideal).
**color**: The diamond color, ranging from J (worst) to D (best).
**clarity**: A measure of diamond purity (I1, SI2, SI1, VS2, VS1, VVS2, VVS1, IF)
**depth**: Total depth percentage, calculated as $depth = \frac{2z}{x + y}$.
**table:** The width of the diamond's top (the table) relative to its widest point.
**price**: The price of the diamond in US dollars.
**x**: Length of the diamond in mm.

**y**: Width of the diamond in mm.

**z**: Depth of the diamond in mm.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Identify numerical columns automatically
numerical_cols = df.select_dtypes(include=['number']).columns

# 2. Set up the figure and axes
# We calculate rows based on number of columns to keep it dynamic
n_cols = len(numerical_cols)
fig, axes = plt.subplots(nrows=(n_cols + 2) // 3, ncols=3, figsize=(15, 10))
axes = axes.flatten()

# 3. Loop through and create distribution plots
for i, col in enumerate(numerical_cols):
    sns.histplot(df[col], kde=True, ax=axes[i], color='skyblue')
    axes[i].set_title(f'Distribution of {col}', fontsize=12)
    axes[i].set_xlabel('')

# 4. Remove any empty subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 1. Calculate the Interquartile Range (IQR)
Q1 = df['carat'].quantile(0.25)
Q3 = df['carat'].quantile(0.75)
IQR = Q3 - Q1

# 2. Define the outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 3. Filter the dataframe
# This keeps only the data within the "whiskers"
df= df[(df['carat'] >= lower_bound) & (df['carat'] <= upper_bound)]
```
