# one hot encoder

```python
df=pd.DataFrame({"animals":['cat','dog','fox']})
print(df)
```

```python
encoded_df=pd.get_dummies(df,columns=['animals'],dtype=int)
print(encoded_df)
```

```python
#sklearn
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(sparse_output=False)
dff=ohe.fit_transform(df)
print(dff)
```

```python
df.job_type.unique()
```

```python
en=pd.get_dummies(df['job_type'],prefix='job_type', dtype=int)
print(en)
```

```python
df=pd.concat([df,en],axis=1)
df
```

```python

encode=ohe.fit_transform(df[['job_type']])
encode
```

```python
en_dff=pd.DataFrame(encode)
```

```python
df=pd.concat([df,en_dff],axis=1)
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
education=[['ssc'],['pg'],['ug'],['inter']]
oe=OrdinalEncoder(categories=[['ssc','inter','ug','pg']])
ev=oe.fit_transform(education)
ev
```

```python
df.social_platform_preference.unique()
```

```python
e=OrdinalEncoder(categories=[['Instagram','TikTok','Telegram','Twitter','Facebook']])
v=e.fit_transform(df[['social_platform_preference']])
v
```

```python
enf=pd.DataFrame(v)
```

```python
df=pd.concat([df,enf],axis=1)
df
```

# Column Transformer

```python
num_clos=df.select_dtypes(include="number").columns
cat_clos=df.select_dtypes(include="object").columns
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder , StandardScaler
```

```python
pre =ColumnTransformer(transformers=[("ohe",OneHotEncoder(),cat_clos),("sc",StandardScaler(),num_clos)])
```

```python
prep=pre.fit_transform(df)
prep
```

```python
dfff=pd.DataFrame(prep)
dfff
```
