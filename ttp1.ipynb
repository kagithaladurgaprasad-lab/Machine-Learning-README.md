```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
corpus=["I love NLP","I love machine learning ","NLP is usefull in data science",
        "Machine learning is part of data science",
        "I enjoy learning NLP"]
```

```python
vect=CountVectorizer(binary=True)
```

```python
bag=vect.fit_transform(corpus)
```

```python
print("Vocabulary:",vect.get_feature_names_out())
print("Bow Matrix:\n", bag.toarray())
```

```python
from collections import Counter
import re

#Function to clean text and extract words
def get_words(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r"[^a-zA-Z\s]", "", text)  # Remove special characters
    words = text.split()
    return words
#Get all words from the dataset
all_words = df["cleaned_review"].apply(get_words).sum()
#Count word frequencies
word_counts = Counter(all_words)
#Display the most common words
print(word_counts.most_common(10))
!pip install wordcloud
```
