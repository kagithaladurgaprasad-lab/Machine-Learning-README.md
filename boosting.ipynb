# AdaBoostClassifier

```python
from sklearn.datasets import load_iris
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report,r2_score,mean_squared_error
```

```python
iris=load_iris()
X,y=iris.data,iris.target
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
base_estimator=DecisionTreeClassifier(max_depth=1)
```

```python
ada_clf=AdaBoostClassifier(estimator=base_estimator,n_estimators=50,learning_rate=1.0,random_state=42)
ada_clf.fit(X_train,y_train)
```

```python
#predictions and evaluation 
y_pred=ada_clf.predict(X_test)
print("Accuracy:",accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

## GradientBoostingRegressor

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
#load dataset
cal=fetch_openml(name='california_housing',version=1,as_frame=True)
x=cal.data
y=cal.target
```

```python
x.drop('ocean_proximity',axis =1 ,inplace=True)
```

```python
x['total_bedrooms']=x['total_bedrooms'].fillna(value=537)
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
gbr = GradientBoostingRegressor(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=3,
    random_state=42
)
```

```python
# Train model
gbr.fit(X_train, y_train)

# Predictions
y_pred = gbr.predict(X_test)

# Evaluation
print("R2 Score:", r2_score(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
```

```python
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score,classification_report,r2_score,mean_squared_error

iris=load_iris()
X,y=iris.data,iris.target

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```
