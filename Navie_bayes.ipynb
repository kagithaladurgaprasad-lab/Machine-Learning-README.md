```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("banana_quality_dataset.csv")
```

```python
#drop 
df.drop(["sample_id","harvest_date"],axis=1,inplace=True) 
```

```python
y=df['quality_category']
```

```python
x =df.drop(['quality_category'],axis=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder,StandardScaler
ct=ColumnTransformer(transformers=[("oe",OrdinalEncoder(),[0,1,4]),
                                   ("sc",StandardScaler(),[2,3,5,6,7,8,9,10,11,12])])
```

```python
x_transformed=ct.fit_transform(x)
```

```python
x_transformed=pd.DataFrame(x_transformed,columns=x.columns)
x_transformed
```

```python
#feature selection
from sklearn.feature_selection import mutual_info_classif
mf=mutual_info_classif(x_transformed,y)
```

```python
mf=pd.Series(mf)
mf.index=x_transformed.columns
mf.sort_values(ascending=False)
```

```python
x.drop(["firmness_kgf","variety","weight_g","altitude_m","soil_nitrogen_ppm"],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
ct=ct=ColumnTransformer(transformers=[("oe",OrdinalEncoder(),[0,3,]),
                                   ("sc",StandardScaler(),[1,2,4,5,6,7])])
```

```python
x_train_transform=ct.fit_transform(x_train)
x_train_transform=pd.DataFrame(x_train_transform)
```

```python
x_test_transform=ct.transform(x_test)
x_test_transform=pd.DataFrame(x_test_transform)
```

```python
from sklearn.naive_bayes import GaussianNB
gn=GaussianNB()
gn.fit(x_train_transform,y_train)
y_pred=gn.predict(x_train_transform)
```

```python
from sklearn.metrics import f1_score
f1=f1_score(y_train,y_pred,average='micro')
f1
```
