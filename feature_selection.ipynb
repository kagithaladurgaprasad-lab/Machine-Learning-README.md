```python
import pandas as pd
```

```python
df=pd.read_csv("banana_quality_dataset.csv")
df.head()
```

```python
y=df['quality_category']
```

```python
x=df.drop(columns=["quality_category","sample_id","harvest_date"],axis=1)
```

```python
from sklearn.preprocessing  import OrdinalEncoder,MinMaxScaler
from sklearn.compose import ColumnTransformer
transformer=ColumnTransformer(transformers=[("t1",OrdinalEncoder(),[0,1,4]),
                                            ("t2",MinMaxScaler(),[2,3,5,6,7,8,9,10,11,12])])

X_transformed=transformer.fit_transform(x)
```

```python
X_transformed=pd.DataFrame(X_transformed,columns=x.columns)
```

```python
X_transformed.head()
```

```python
#matual info classifer
from sklearn.feature_selection import mutual_info_classif
mutual_info=mutual_info_classif(X_transformed,y)
mutual_info=pd.Series(mutual_info)
mutual_info.index=X_transformed.columns
mutual_info.sort_values(ascending=False)
```

```python
from sklearn.feature_selection import chi2
import pandas as pd
#Apply chi-square test
chi_scores,_=chi2(X_transformed,y)

#create DataFrame with only chi-suARE scores 
chi_df=pd.DataFrame({
    'feature':X_transformed.columns,
    'chi2_score':chi_scores
})
#sort features ny importance 
chi_df=chi_df.sort_values(by='chi2_score',ascending=False)
print(chi_df)
```

```python
from sklearn.feature_selection import chi2,SelectKBest
select_columns=SelectKBest(chi2,k=5)
best_features=select_columns.fit_transform(X_transformed,y)
```

```python
selected_features=x.columns[select_columns.get_support()]
print(selected_features)
```
