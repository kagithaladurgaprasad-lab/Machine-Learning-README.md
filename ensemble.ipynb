```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
iris=load_iris()
x=iris.data
y=iris.target

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
mod1=LogisticRegression(random_state=42)
mod2=DecisionTreeClassifier(random_state=42)
mod3=SVC(probability=True,random_state=42)
```

```python
voting_clf=VotingClassifier(estimators=[
    ('lr',mod1),
    ('dt',mod2),
    ('svc',mod3)
],voting='soft')
```

```python
voting_clf.fit(x_train,y_train)

#predict
y_pred=voting_clf.predict(x_test)
acc=accuracy_score(y_test,y_pred)

acc
```

```python
from sklearn.ensemble import StackingClassifier
```

```python
base_learners=[('dt',DecisionTreeClassifier()),
               ('svc',SVC(probability=True))]
meta_model=LogisticRegression()
stack=StackingClassifier(estimators=base_learners,final_estimator=meta_model,cv=5)
stack.fit(x_train,y_train)
print(stack.score(x_test,y_test))
```
