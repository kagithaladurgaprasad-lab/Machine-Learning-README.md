# part 4 & part 5 :-----Data Scaling & visualizing before and after the scaling

```python
# original distribution before applying scaling
```

```python
df.hist(figsize=(12,10), bins=20)
plt.suptitle("Original Distribution")
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

minmax = scaler.fit_transform(df)
```

```python
minmax_scaled_df=pd.DataFrame(minmax, columns=df.columns) 
minmax_scaled_df
```

```python
minmax_scaled_df.hist(figsize=(12,10), bins=20)
plt.suptitle("Original Distribution")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
ss = scaler.fit_transform(df)
```

```python
standard_scaled_df=pd.DataFrame(ss, columns=df.columns) 
standard_scaled_df
```

```python
standard_scaled_df.hist(figsize=(12,10), bins=20)
plt.suptitle("Plot Distribution")
plt.show()
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
rs = scaler.fit_transform(df)
```

```python
robust_scaled_df=pd.DataFrame(rs, columns=df.columns) 
robust_scaled_df
```

```python
robust_scaled_df.hist(figsize=(12,10), bins=20)
plt.suptitle("Plot Distribution")
plt.show()
```

```python
## 3 


#How did scaling reshape the distribution?
#ans:-- Scaling did not change the shape of the distribution, but it changed the scale (range and spread) of values.

#Which scaler handled outliers most effectively?
#ans:-- robust scaler because it takes median and IQR , and ignores extreme values

#How MinMaxScaler affects feature ranges
#ans:-- MinMaxScaler transforms all values into a fixed range (0 to 1)

#How StandardScaler affects mean and variance
#ans:-- StandardScaler standardizes features by centering them around a mean of zero and scaling to unit variance,
#but it is sensitive to outliers since it relies on mean and standard deviation.
```
