```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# load dataset
df=pd.read_csv("loan_approved.csv")
df.head()
```

```python
df.info()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
num_cols = df.select_dtypes(include='number').columns
cat_cols = df.select_dtypes(include='object').columns

print("Numerical columns:", num_cols)
print("Categorical columns:", cat_cols)
```

```python
for col in num_cols:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
df.info()
```

```python
y=df["Loan_Status (Approved)"]
```

```python
## input varibles 13 columns
x=df.drop(["Loan_Status (Approved)"],axis=1)
```

```python
x.info()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder,StandardScaler
ct=ColumnTransformer(transformers=[("oe",OrdinalEncoder(),[0,1,2,3,4,5,11]),
                                   ("sc",StandardScaler(),[6,7,8,9,10])])
```

```python
x_transformed=ct.fit_transform(x)
```

```python
x_transformed
```

```python
x_transformed=pd.DataFrame(x_transformed,columns=x.columns)
x_transformed
```

```python
from sklearn.feature_selection import mutual_info_classif
mf=mutual_info_classif(x_transformed,y)
mf=pd.Series(mf)
mf.index=x_transformed.columns
mf.sort_values(ascending=False)
```

```python
x.drop(["Loan_Amount_Term","Self_Employed","Dependents","Married","Loan_ID","CoapplicantIncome"], axis=1,inplace=True)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x.info()
```

```python
ct=ColumnTransformer(transformers=[("oe",OrdinalEncoder(),[0,1,5]),
                                   ( "sc",StandardScaler(),[2,3,4])])
```

```python
x_train_transform=ct.fit_transform(x_train)
x_train_transform=pd.DataFrame(x_train_transform)
x_train_transform
```

```python
x_test_transform=ct.fit_transform(x_test)
x_test_transform=pd.DataFrame(x_test_transform)
```

```python
from imblearn.over_sampling import SMOTE

smote=SMOTE()
x_smote,y_smote=smote.fit_resample(x_train_transform,y_train)
```

```python
from collections import Counter
print(Counter(y_train))
print(Counter(y_smote))
```

```python
from sklearn.svm import SVC 
svclassifier=SVC()
svclassifier.fit(x_smote,y_smote)
```

```python
y_pred=svclassifier.predict(x_test_transform)
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,classification_report,f1_score
acc=accuracy_score(y_test,y_pred)
```

```python
acc
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred, pos_label='Y')

print(f1)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'C':[0.5,5,10,50,60,70],
            'gamma':[1,0.1,0.01,0.001,0.0001],
            'random_state':(list(range(1,20)))}

model=SVC()
grid=GridSearchCV(model,param_grid,refit=True,verbose=1,scoring='f1',cv=5)

grid.fit(x_smote,y_smote)
```

```python
print(grid.best_params_)
```

```python
from sklearn.svm import SVC 
svclassifier=SVC(C=0.5,gamma=1,random_state=1)
svclassifier.fit(x_smote,y_smote)
```

```python
y_pred=svclassifier.predict(x_test_transform)
```
