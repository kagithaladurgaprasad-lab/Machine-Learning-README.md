```python
#drop 
df.drop(["sample_id","harvest_date"],axis=1,inplace=True)      
```

```python
 #target variable quality category
y=df['quality_category']
```

```python
x =df.drop(['quality_category'],axis=1)
```

```python
x.shape
```

```python
#apply columntransformer 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder,StandardScaler
ct=ColumnTransformer(transformers=[("oe",OrdinalEncoder(),[0,1,4]),
                                   ("sc",StandardScaler(),[2,3,5,6,7,8,9,10,11,12])])
```

```python
x_transformed=ct.fit_transform(x)
```

```python
x_transformed
```

```python
x_transformed=pd.DataFrame(x_transformed,columns=x.columns)
x_transformed
```

```python
#feature selection
from sklearn.feature_selection import mutual_info_classif
mf=mutual_info_classif(x_transformed,y)
```

```python
mf=pd.Series(mf)
mf.index=x_transformed.columns
mf.sort_values(ascending=False)
```

```python
x.drop(["firmness_kgf","variety","weight_g","altitude_m","soil_nitrogen_ppm"],axis=1,inplace=True)
```

```python
x.shape
```

```python
x.info()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
ct=ct=ColumnTransformer(transformers=[("oe",OrdinalEncoder(),[0,3,]),
                                   ("sc",StandardScaler(),[1,2,4,5,6,7])])
```

```python
x_train_transform=ct.fit_transform(x_train)
x_train_transform=pd.DataFrame(x_train_transform)
```

```python
x_test_transform=ct.transform(x_test)
x_test_transform=pd.DataFrame(x_test_transform)
```

```python
#model bulding
from sklearn.neighbors import KNeighborsClassifier
kn=KNeighborsClassifier(n_neighbors=7,metric="minkowski",p=1)
```

```python
#training
kn.fit(x_train_transform,y_train)
```

```python
#test
y_pred=kn.predict(x_test_transform)
```

```python
#evolution
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
ac=accuracy_score(y_test,y_pred)
ac
```

```python
pr=precision_score(y_test,y_pred,average="micro")
pr
```

```python
f1=f1_score(y_test,y_pred,average="micro")
f1
```

```python
from sklearn.metrics import confusion_matrix,classification_report
cm=confusion_matrix(y_test,y_pred)
cm
```
