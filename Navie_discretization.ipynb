# Equal width Binning

```python
# implementation of Equal width binning 
from sklearn.preprocessing import KBinsDiscretizer

#let us create continious data 
import numpy as np
x_conti = np.arange(1,101).reshape(10,10)
print(x_conti)
print("\n")
#using uniform strategy for equal-width bins 
discretizer =KBinsDiscretizer(n_bins=4,encode='ordinal',strategy='uniform')
x_binned=discretizer.fit_transform(x_conti)
x_binned
```

# Equal Frequency Binning

```python
# implementation of Equal frequency  binning 
from sklearn.preprocessing import KBinsDiscretizer

#let us create continious data 
import numpy as np
x_conti = np.arange(1,101).reshape(10,10)
print(x_conti)
print("\n")
#using quantile uniform for equal frequency bins 
discretizer =KBinsDiscretizer(n_bins=4,encode='ordinal',strategy='quantile')
x_binned1=discretizer.fit_transform(x_conti)
x_binned1
```

# k-means

```python
x_conti = np.arange(1,101).reshape(10,10)
print(x_conti)
print("\n")
#using k-means 
discretizer =KBinsDiscretizer(n_bins=4,encode='ordinal',strategy='kmeans')
x_binned2=discretizer.fit_transform(x_conti)
x_binned2
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import KBinsDiscretizer
from sklearn.datasets import load_iris
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score ,classification_report 
```

```python
#load dataset 
data=load_iris()
x=pd.DataFrame(data.data,columns=data.feature_names)
y=data.target

#function to apply discretization
def discretize_data(x,strategy):
    kb=KBinsDiscretizer(n_bins=5,encode='ordinal',strategy=strategy)
    x_binned=kb.fit_transform(x)
    return pd.DataFrame(x_binned,columns=x.columns)

#compare stratigies
strategy=['uniform','quantile','kmeans']
results={}

for st in strategy:
    x_binned=discretize_data(x,st)
    x_train,x_test,y_train,y_test=train_test_split(x_binned,y,test_size=0.3,random_state=42)
    model=MultinomialNB()
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)

    acc=accuracy_score(y_test,y_pred)
    results[st]=acc
    print(f"strategy:{st}")
    print(classification_report(y_test,y_pred))

#plot results
plt.bar(results.keys(),results.values(),color='skyblue')
plt.title("Navie Bayes Accuracy by Discretization Strategy")
plt.ylabel("Accuracy")
plt.show()
```
